```python
import torch
import matplotlib.pyplot as plt
from MolDAIS.MolDAIS import MolDAIS
# from MolDAIS import MolDAIS


# Step 1: Define mock data
smiles_list = [
    "CCO", "CCC", "CCN", "CCCl", "CCCC", "CCCO", "CCCN", "ClCCCl",
    "CCOCC", "CCON", "CCCNCl", "CCCCCl", "CCCCN", "CCCCCO", "CCNCl",
    "CCNO", "ClCCl", "CCOO", "CCNN", "COCO"
]

# Mock logP values for the molecules (as the target objective values)
logP_values = torch.tensor([0.5, 1.2, 0.8, 2.1, 3.0, 1.5, 2.3, 2.8, 0.9, 1.4,
                            2.5, 1.9, 1.1, 2.7, 2.6, 1.6, 1.8, 2.2, 1.3, 1.7], dtype=torch.float32)



# Step 2: Define the problem
problem = MolDAIS.Problem(smiles_search_space=smiles_list,
                          targets=logP_values.unsqueeze(1),
                          experiment_name="LogP_Optimization_Test")
problem.compute_descriptors()

# Step 3: Define optimizer parameters
optimizer_parameters = MolDAIS.OptimizerParameters(

    sparsity_method='MI',  # Use MI as the sparsity method
    acq_fun='EI',  # Use Expected Improvement (EI) as the acquisition function
    num_sparsity_feats=10,  # Let's use 10 features for sparsity

    #one or both must be false
    multi_objective=False,  # Single-objective optimization
    constrained=False,  # No constraints
    #If either are true, acq_fun is overridden

    #should change to budget
    total_sample_budget=7,  # must be greater than initialization budget
    initialization_budget=2,  # Number of initial points
    seed=123  # Seed for reproducibility
)

# Step 4: Create the optimization problem and the MolDAIS object
mol_dais = MolDAIS(problem=problem, optimizer_parameters=optimizer_parameters)


# Step 5: Run the optimization for a few iterations
print("Running optimization for the first time:")
mol_dais.configuration.optimize()

print(mol_dais.results.best_molecules)
print(mol_dais.results.best_values)

# Step 7: Plot the convergence plot
mol_dais.configuration.plot_convergence()


```

```python

# Step 6: Save a checkpoint and reload it to continue the optimization
# since the last run completed, this should do nothing
checkpoint_filename = 'LogP_Optimization_Test_sparsity-MI_model-GP_acq-EI_iter_5_seed-123.pkl'
print("\nRestarting from checkpoint:")
mol_dais.configuration.load_from_checkpoint(checkpoint_filename)

mol_dais.configuration.optimize()
mol_dais.configuration.plot_convergence()
```

```python

# Step 6: Save a checkpoint and reload it to continue the optimization
#
checkpoint_filename = 'LogP_Optimization_Test_sparsity-MI_model-GP_acq-EI_iter_5_seed-123.pkl'
print("\nRestarting from checkpoint:")
mol_dais.configuration.load_from_checkpoint(checkpoint_filename)


# incress iterations to coniniue where we left off
mol_dais.optimizer_parameters.iterations = 12
mol_dais.optimizer_parameters.initialization_budget = 0
mol_dais.configuration.optimize()
mol_dais.configuration.plot_convergence()
```

### Simple Multi-objective

```python
import torch
import matplotlib.pyplot as plt

# Step 1: Define mock data
smiles_list = ["CCO", "CCC", "CCN", "CCCl", "CCCC", "CCCO", "CCCN", "ClCCCl",
              "CCOCC", "CCON", "CCCNCl", "CCCCCl", "CCCCN", "CCCCCO", "CCNCl",
              "CCNO", "ClCCl", "CCOO", "CCNN", "COCO"]
# Mock logP values for the molecules (as the target objective values)
two_target_values = torch.tensor([[0.5, 1.2, 0.8, 2.1, 3.0, 1.5, 2.3, 2.8, 0.9, 1.4,
                            2.5, 1.9, 1.1, 2.7, 2.6, 1.6, 1.8, 2.2, 1.3, 1.7],
                              [0.5, 1.2, 0.8, 2.1, 3.0, 1.5, 2.3, 2.8, 0.9, 1.4,
                            2.5, 1.9, 1.1, 2.7, 2.6, 1.6, 1.8, 2.2, 1.3, 1.7]]
                              , dtype=torch.float32).T
two_target_values[:,1] = torch.abs(6/(2-two_target_values[:,1])) #= two_target_values[:,1]


# Step 2: Define the problem
problem = MolDAIS.Problem(smiles_search_space=smiles_list,
                          targets=two_target_values,
                          experiment_name="MO_LogP_Optimization_Test")
problem.compute_descriptors()

# Step 3: Define optimizer parameters
optimizer_parameters = MolDAIS.OptimizerParameters(
    sparsity_method='MI',  # Use MIC as the sparsity method
    acq_fun='EI',  # Use Expected Improvement (EI) as the acquisition function
    num_sparsity_feats=10,  # Let's use 10 features for sparsity
    multi_objective=True,  # Single-objective optimization
    constrained=False,  # No constraints
    iterations=7,  # Number of iterations to run
    initialization_budget=2,  # Number of initial points
    seed=12  # Seed for reproducibility
)

# Step 4: Create the optimization problem and the MolDAIS object
mol_dais = MolDAIS(problem=problem, optimizer_parameters=optimizer_parameters)


# Step 5: Run the optimization for a few iterations
print("Running optimization for the first time:")
mol_dais.configuration.optimize()


# Step 7: Plot the convergence plot
mol_dais.configuration.plot_convergence()
print(mol_dais.results.best_molecules)
print(mol_dais.results.best_values)


```

### Simple batch

```python
import torch
import matplotlib.pyplot as plt

# Step 1: Define mock data
smiles_list = [
    "CCO", "CCC", "CCN", "CCCl", "CCCC", "CCCO", "CCCN", "ClCCCl",
    "CCOCC", "CCON", "CCCNCl", "CCCCCl", "CCCCN", "CCCCCO", "CCNCl",
    "CCNO", "ClCCl", "CCOO", "CCNN", "COCO"
]
# Mock logP values for the molecules (as the target objective values)
logP_values = torch.tensor([0.5, 1.2, 0.8, 2.1, 3.0, 1.5, 2.3, 2.8, 0.9, 1.4,
                            2.5, 1.9, 1.1, 2.7, 2.6, 1.6, 1.8, 2.2, 1.3, 1.7], dtype=torch.float32)



# Step 2: Define the problem
problem = MolDAIS.Problem(smiles_search_space=smiles_list,
                          targets=logP_values.unsqueeze(1),
                          experiment_name="LogP_Optimization_Test")
problem.compute_descriptors()

# Step 3: Define optimizer parameters
optimizer_parameters = MolDAIS.OptimizerParameters(
    sparsity_method='MIC',  # Use MIC as the sparsity method
    acq_fun='EI',  # Use Expected Improvement (EI) as the acquisition function
    num_sparsity_feats=10,  # Let's use 10 features for sparsity
    multi_objective=False,  # Single-objective optimization
    constrained=False,  # No constraints
    iterations=7,  # Number of iterations to run
    initialization_budget=2,  # Number of initial points
    seed=123,   # Seed for reproducibility
    batch = 3 # batch size
)

# Step 4: Create the optimization problem and the MolDAIS object
mol_dais = MolDAIS(problem=problem, optimizer_parameters=optimizer_parameters)


# Step 5: Run the optimization for a few iterations
print("Running optimization for the first time:")
mol_dais.configuration.optimize()

print(mol_dais.results.best_molecules)
print(mol_dais.results.best_values)

# Step 7: Plot the convergence plot
mol_dais.configuration.plot_convergence()


```

### Photoswitch test

```python
from gauche.dataloader import MolPropLoader

# load a benchmark dataset
loader = MolPropLoader()
loader.load_benchmark("Photoswitch")

import torch
import matplotlib.pyplot as plt

# Step 1: load data
smiles_list = loader.features
values = torch.tensor(loader.labels, dtype=torch.float32)

values.shape, len(smiles_list)


# Step 2: Define the problem
PS_problem = MolDAIS.Problem(smiles_search_space=smiles_list,
                          targets=values,
                          experiment_name="Photoswitch_Optimization_Test")
PS_problem.compute_descriptors()


# Step 3: Define optimizer parameters
optimizer_parameters = MolDAIS.OptimizerParameters(
    sparsity_method='MIC',  # Use MIC as the sparsity method
    acq_fun='UCB',  # Use Expected Improvement (EI) as the acquisition function
    num_sparsity_feats=10,  # Let's use 10 features for sparsity
    multi_objective=False,  # Single-objective optimization
    constrained=False,  # No constraints
    iterations=15,  # Number of iterations to run
    initialization_budget=5,  # Number of initial points
    seed=1234  # Seed for reproducibility
)

# Step 4: Create the optimization problem and the MolDAIS object
mol_dais = MolDAIS(problem=PS_problem, optimizer_parameters=optimizer_parameters)


# Step 5: Run the optimization for a few iterations
print("Running optimization for the first time:")
mol_dais.optimization_problem.optimize()

print(mol_dais.optimization_problem.best_molecules)
print(mol_dais.optimization_problem.best_values)

# Step 7: Plot the convergence plot
mol_dais.optimization_problem.plot_convergence()


```
